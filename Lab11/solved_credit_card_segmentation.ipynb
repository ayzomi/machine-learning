# Easy Credit Card Customer Segmentation Solution

## 1) Import Libraries

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
```

## 2) Read Dataset

```
df = pd.read_csv('CC_GENERAL.csv')
```

## 3) Show First Rows

```
df.head()
```

## 4) Dataset Shape

```
df.shape
```

## 5) Dataset Information

```
df.info()
```

## 6) Summary Statistics

```
df.describe()
```

## 7) Drop Customer ID

```
df = df.drop('CUST_ID', axis=1)
```

## 8) Check Missing Values

```
df.isnull().sum()
```

## 9) Fill Missing Values

```
df = df.fillna(df.mean())
```

## 10) Histograms

```
df.hist(figsize=(20,20))
plt.show()
```

Simple explanation: Histograms show how the values are distributed.

## 11) Correlation Heatmap

```
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(), cmap='coolwarm')
plt.show()
```

Simple explanation: The heatmap shows strong and weak relationships between columns.

## 12) Scatter Plot

```
plt.figure(figsize=(8,6))
plt.scatter(df['BALANCE'], df['PURCHASES'])
plt.xlabel('BALANCE')
plt.ylabel('PURCHASES')
plt.title('BALANCE vs PURCHASES')
plt.show()
```

## 13) Feature Scaling

```
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

Simple explanation: Scaling makes all features on the same level before clustering.

## 14) Elbow Method

```
inertia_values = []

for k in range(1,11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(scaled_data)
    inertia_values.append(kmeans.inertia_)
```

## 15) Plot Elbow Curve

```
plt.figure(figsize=(8,6))
plt.plot(range(1,11), inertia_values, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

Simple explanation: We choose the K value near the elbow point.

## 16) Silhouette Scores

```
silhouette_scores = []

for k in range(2,11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(scaled_data)
    score = silhouette_score(scaled_data, labels)
    silhouette_scores.append(score)
```

## 17) Plot Silhouette Scores

```
plt.figure(figsize=(8,6))
plt.plot(range(2,11), silhouette_scores, marker='o')
plt.xlabel('K Value')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores')
plt.show()
```

Simple explanation: Higher silhouette score means better clusters.

## 18) Final K-Means Model

```
final_kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
final_labels = final_kmeans.fit_predict(scaled_data)
```

## 19) Add Cluster Labels

```
df['Cluster'] = final_labels
df.head()
```

## 20) Cluster Summary

```
cluster_summary = df.groupby('Cluster').mean()
cluster_summary
```

## 21) Number of Customers in Each Cluster

```
df['Cluster'].value_counts()
```

## 22) PCA Visualization

```
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)

plt.figure(figsize=(8,6))
plt.scatter(pca_data[:,0], pca_data[:,1], c=final_labels)
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('Customer Clusters')
plt.show()
```

# Final Answers

### Why is this unsupervised learning?
Because there are no target labels.

### Why did we remove CUST_ID?
Because it is only an ID and does not help clustering.

### Why is scaling important?
Because K-Means depends on distance calculations.

### What do the clusters represent?
Groups of customers with similar spending behavior.

### Why did we use PCA?
To visualize the clusters in 2D.
